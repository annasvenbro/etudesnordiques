# 1. Import des paquets nécessaires

```python
import requests as rq
import pandas as pd
```

# 2. Requête de l'API IdRef

```python
url = "https://www.idref.fr/services/listrcr"
response = rq.get(url)
response.raise_for_status()  # arrêt si erreur réseau
```

# 3. Transformation en tableau

```python
lines = response.text.split("\n")
header = lines[0].split("\t")
header[0] = header[0].strip("\ufeff")  # nettoyage BOM
data = [line.split("\t") for line in lines[1:] if line]
df_rcr = pd.DataFrame(data, columns=header)
```

# 4. Nettoyage des colonnes importantes

```python
df_rcr.columns = df_rcr.columns.str.strip("\ufeff")
df_rcr["RCR"] = (
    df_rcr["RCR"].str.replace("=", "").str.replace('"', "")
)
df_rcr["PPN"] = (
    df_rcr["PPN"].str.replace("=", "").str.replace('"', "")
)
df_rcr = df_rcr.rename(columns={"LONGITUDE\r": "LONGITUDE"})
df_rcr["LONGITUDE"] = df_rcr["LONGITUDE"].str.rstrip("\r")
```

# 5. On ne garde que la colonne RCR

```python
df_rcr = df_rcr.filter(regex="^RCR$")
```
